```python
import sympy as sym
from IPython.display import display,Math
sym.init_printing


x = sym.symbols('x')
mu, alpha, sigma = sym.symbols('mu,alpha,sigma')

#exp stands for expression
expr = sym.exp((mu-alpha)**2 / (2*sigma**2))

display(expr)
```

```python
x,y = sym.symbols('x,y')
expr = x+4 *y
display(expr)

#Subsitution
expr.subs(x,6)
#6 was substituted in for x in the expression

#use a dictionary to substitute in multiple variables
expr.subs({x:3, y:3})
```

```python
expr = 3/x

#sym.latex will output the latex code for the expression
display(sym.latex(expr))

display(Math(sym.latex(expr)))

#convert what would produce a value into a sympy object
display(sym.latex(sym.sympify('3/4')))


expr = x**2 + 4

for i in range(-2,3):
    ans = (x+4).subs(x, i**2)
    display(Math('\\text{"With "}x=%g, x^2+4 \\quad \\Rightarrow \\quad %g^2+4= %g'%(i,i,ans)))
```
